```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

# Functions

```python
# objective function
def f(x):
    #return x**2 + 1
    #return x**4 - 4*x**3 - 12*x**2 + 16*x + 5
    return ( (x-1)**2+0.15 )*( (x+1)**2 )

# derivative/gradient of objective function
def df(x):
    #return 2*x
    #return 4*x**3 - 12*x**2 - 24*x + 16
    return 2*(x - 1)*((x + 1)**2) + ((x - 1)**2 + 0.15)*(2*(x + 1))

# gradient descent function
def GD(df,start,rate,nStep):
    # init points array
    points = np.zeros(nStep)
    
    # init starting guess
    points[0] = start

    # run GD loop
    for i in range(nStep-1):
         points[i+1] = points[i] - rate*df(points[i])
    # end
    
    return points
# end
```

# Run gradient descent

```python
nStep = 100  # number of iterations/steps
start = -3   # initial guess
rate = 0.05  # learning rate

# run GD
points = GD(df,start,rate,nStep)

# get results
arg_min = np.argmin( f(points) )
x_min = points[arg_min]
f_min = f( x_min )
print(f"Minimum x: {x_min}")
print(f"Minimum f: {f_min}")

# plot history
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12,3))

graphX = np.linspace(-start*2,start*2,100)

axes[0].plot(graphX,f(graphX),'r--')
axes[0].plot(points,f(points))
axes[0].set_xlabel("x")
axes[0].set_ylabel("f(x)")

axes[1].plot(points)
axes[1].set_xlabel("step")
axes[1].set_ylabel("x")

axes[2].plot(f(points))
axes[2].set_xlabel("step")
axes[2].set_ylabel("f(x)")

plt.show()
```
