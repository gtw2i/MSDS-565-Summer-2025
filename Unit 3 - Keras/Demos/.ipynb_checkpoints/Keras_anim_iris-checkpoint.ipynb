### Dimensionality reduction

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import (FactorAnalysis, FastICA, IncrementalPCA, KernelPCA, PCA, TruncatedSVD)
from sklearn.manifold import (Isomap, MDS, SpectralEmbedding, TSNE)
from matplotlib import gridspec

# load and scale iris data
df = sns.load_dataset('iris')
X = df.drop('species', axis=1).to_numpy()
y = df['species'].to_numpy()
sc = StandardScaler()
X_sc = sc.fit_transform(X)

# encode species names into ints
le = LabelEncoder()
y_enc = le.fit_transform(y)

decomps = [
    # decomposition
    FactorAnalysis(n_components=2),
    FastICA(n_components=2),
    IncrementalPCA(n_components=2),
    KernelPCA(n_components=2, kernel='rbf'),
    PCA(n_components=2),
    TruncatedSVD(n_components=2),
    # manifold
    Isomap(n_components=2),
    MDS(n_components=2),
    SpectralEmbedding(n_components=2),
    TSNE(n_components=2)
]

def plot_dimensionality_reductions(X, y_encoded, decomps):
    ncols = 5  # Number of columns per row
    nrows = (len(decomps) + ncols - 1) // ncols  

    fig = plt.figure(figsize=(5 * ncols, 4 * nrows)) # size of figure
    gs = gridspec.GridSpec(nrows, ncols)

    for i, model in enumerate(decomps):
        ax = fig.add_subplot(gs[i // ncols, i % ncols])
        X_transformed = model.fit_transform(X)
        scatter = ax.scatter(X_transformed[:, 0], X_transformed[:, 1], c=y_encoded, cmap='jet', s=10)
        plt.title(str(model), fontsize=14)
        plt.colorbar(scatter, ax=ax, ticks=range(len(le.classes_)), label='Species')

    plt.suptitle("Different dimensionality reduction techniques", fontsize=18, y=1.02)
    plt.tight_layout()
    plt.show()

plot_dimensionality_reductions(X_sc, y_enc, decomps)
```

```python
# load iris data

df = sns.load_dataset('iris')
X = df.drop('species', axis=1).to_numpy()
y = df['species'].to_numpy()

sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# choose a dimensionality reduction technique

pca = PCA(n_components=2)
X = pca.fit_transform(X)

# tsne  =  TSNE(n_components=2)  # not sure why but TSNE does not plot correctly 
# X = tsne.fit_transform(X)

# fa  =  FactorAnalysis(n_components=2)
# X = fa.fit_transform(X)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

OHE = OneHotEncoder()
y = OHE.fit_transform(y.reshape(-1, 1)).toarray()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    verbose=1)
```

```python
# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
y_test_prob = model.predict(X_test)
y_test_prob
```

```python
y_test_pred = y_test_prob.argmax(axis=1)
y_test_pred
```

```python
from sklearn.metrics import  classification_report

print(classification_report(y_test.argmax(axis=1),
                            y_test_pred))
```

# animate function but with dynamic elements (easier to use on real data)

```python
import matplotlib.animation as animation
%matplotlib widget

def animate(model, X_train, y_train, X_test, y_test, nGrid=200):
    fig, ax = plt.subplots(figsize=(7, 6))
    predicted_frames = []
   
    # calculate range of feature space
    x_min, x_max = np.min(X_train[:, 0])*1.2, np.max(X_train[:, 0])*1.2
    y_min, y_max = np.min(X_train[:, 1])*1.2, np.max(X_train[:, 1])*1.2
    
    # create grid with +1 and -1 margin and nGrid evenly spaced points
    a1 = np.linspace(x_min, x_max, nGrid)
    a2 = np.linspace(y_min, y_max, nGrid)
    A1, A2 = np.meshgrid(a1, a2)
    A = np.vstack( [A1.ravel(), A2.ravel()] ).T  # stack flatten transpose
    extent = [x_min, x_max, y_min, y_max]

    for i in range(40):  # Number of frames in the animation
        model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test), verbose=0)
        B = model.predict(A).argmax(axis=1)
        B = B.reshape(nGrid, nGrid)
        predicted_frames.append(B)

    scatter = ax.scatter(X_train[:, 0], X_train[:, 1], c=np.argmax(y_train, axis=1), cmap='jet', s=5)

    plot = ax.imshow(predicted_frames[0], animated=True, cmap='jet', alpha=0.5, extent=extent) # plot the first frame

    def update(frame):
        plot.set_data(predicted_frames[frame])
        ax.set_title(f"Epoch {(frame+1) * 5}")
        return plot, scatter

    ani = animation.FuncAnimation(fig, update, frames=len(predicted_frames), interval=200, blit=False)
    plt.gca().set_aspect('equal', adjustable='datalim')
    plt.autoscale()
    plt.show()
    return ani

animate(model, X_train, y_train, X_test, y_test)
```
