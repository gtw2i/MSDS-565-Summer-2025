# Animating Keras classification during training
matplotlib technique  https://matplotlib.org/stable/gallery/animation/dynamic_image.html

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

## Create data

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000,    # number of data points
                  centers=3,         # number of clusters
                  cluster_std=1.3,   # std dev (spread) of clusters
                  random_state=0)    

X -= np.mean(X,axis=0)  

%matplotlib inline

cmap = 'jet'
MAX = np.max(np.abs(X))*1.2 
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap); 
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')  
```

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=500,
                  centers=3,
                  cluster_std=0.6,
                  random_state=1)

transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)

X -= np.mean(X,axis=0)

cmap = 'jet'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=500,
                    factor=0.5,
                    noise=0.05)

MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Preprocess

### Train-test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y)
```

# Keras 

### Create model

```python
from keras.models import Sequential  
from keras.layers import Dense   

model = Sequential()

#model.add(Dense(y.shape[1], input_dim=X.shape[1], activation='softmax'))

# Add layers
model.add(Dense(8, input_dim=X.shape[1], activation='relu'))    # first layer (input layer)                                      
model.add(Dense(4, activation='relu'))                          # hidden layer    
model.add(Dense(len(np.unique(y)), activation='softmax'))       # output layer 

# compiling the model: configures the model for training
model.compile(optimizer='adam',    
              loss='sparse_categorical_crossentropy',             
              metrics=['accuracy']) 

model.summary()  
```

```python
# Prepare a spatial grid

# set the grid size (nGrid) and range (a1, a2)
nGrid = 200 
a1 = np.linspace(-MAX,MAX,nGrid)  
a2 = np.linspace(-MAX,MAX,nGrid) 

# create a meshgrid 
A1, A2 = np.meshgrid(a1,a2)
A1 = A1.flatten()    # transform 2D matrix of grid points back into a list format
A2 = A2.flatten()
A = np.vstack((A1,A2)).T   # vertical stack and transpose, so that each row of A represents the (x,y) coords of a point 

# setting plot extent 
extent = [-MAX,MAX,-MAX,MAX] 
```

# FuncAnimation

### animate decision boundaries by pre-computing predictions and then displaying them one by one 

- This method results in a gif-like animation. It runs more smoothly and its frame rate can be customized   
- but it requires waiting time upfront 
- an alternative approach to this could be to display predictions as they come in, but to do so efficiently requires more steps and considerations when coding, such as the incorporation of stoppage rules or pause button

```python
import matplotlib.animation as animation
%matplotlib widget

# # define function: need model, train/test data, grid (A, nGrid)
def animate(model, X_train, y_train, X_test, y_test, A, nGrid):
    fig, ax = plt.subplots(figsize=(7, 6))

    predicted_frames = [] # store processed prediction matrices
    for i in range(40):  # number of frames in animation
        model.fit(X_train, y_train, epochs=1, batch_size=32, validation_data=(X_test, y_test), verbose=0)
        B = model.predict(A).argmax(axis=1)
        B = B.reshape(nGrid, nGrid)
        B = np.flipud(B)
        predicted_frames.append(B) 

    # Plot actual data
    scatter = ax.scatter(X[:, 0], X[:, 1], c=y, s=5, cmap=cmap)

    plot = ax.imshow(predicted_frames[0], animated=True, cmap=cmap, alpha=0.5, extent=extent) # intial frame

    # update function, creates animation when repeatedly called 
    def update(frame):
        plot.set_data(predicted_frames[frame]) # plot current frame's predictions
        ax.set_title(f"Epoch {(frame+1)*5}")
        return plot,  

    # delay between updates of 200 ms; blit=False redraws the plot at each frame
    ani = animation.FuncAnimation(fig, update, frames=len(predicted_frames), interval=200, blit=False) # adjust interval for faster/slower speed
    plt.show()
    return ani,

animate(model, X_train, y_train, X_test, y_test, A, nGrid)
```
