```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm

#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

# Load data

```python
df = pd.read_csv('../data/Real Estate/Real Estate Data.csv', index_col=0, header=0)
df = df.reset_index(drop=True)

df = df.rename(columns = {'Type':'Dwelling Type',
                          'OvQual':'Overall Qual',
                          'Nbhd':'Neighborhood',
                          'Built':'Year Built'})

df.info()
df.head(2)
```

# Fill nans

### check for null numeric columns

```python
gb1 = df.groupby(["Neighborhood"])['Lot Frontage'].median().sort_values()

def fill_1(r):
    n = r.Neighborhood
    
    x = r['Lot Frontage']
    
    if pd.isna(x):
        return gb1[n]
    else:
        return x
    # end
# end

df['Lot Frontage'] = df.apply(fill_1, axis=1)
```

```python
gb2 = df.groupby(["Neighborhood"])['Garage Yr Built'].median().sort_values()

def fill_2(r):
    n = r.Neighborhood
    
    x = r['Garage Yr Built']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Garage Yr Built'] = df.apply(fill_2, axis=1)
```

```python
gb3 = df.groupby(["Neighborhood"])['Masonry/Veneer Area'].median().sort_values()

def fill_3(r):
    n = r.Neighborhood
    
    x = r['Masonry/Veneer Area']
    
    if pd.isna(x):
        return gb2[n]
    else:
        return x
    # end
# end

df['Masonry/Veneer Area'] = df.apply(fill_3, axis=1)
```

```python
df_num = df.select_dtypes(exclude=object)
print(df_num.columns)

df_obj = df.select_dtypes(include=object)
print(df_obj.columns)
```

```python
for c in df_num.columns:
    plt.figure(figsize=(6,3))
    sns.scatterplot( data=df_num, x=c, y='Sale Price')
```

# Setup data

### Get feautres and labels

Use numpy arrays. Keras doesn't like dataframes.

```python
X = df_num.drop('Sale Price', axis=1).to_numpy()
y = df_num['Sale Price'].to_numpy()
y
```

### Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2)
```

# Keras

### Create model

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, input_dim=X.shape[1], activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='relu'))

model.compile(optimizer='adam',
              loss='mean_squared_error')
```

### Train model

```python
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    verbose=1)
```

### Plot training and validation loss

```python
plt.figure(figsize=(10,5))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
```

### Predict model on the test set

```python
y_train_pred = model.predict(X_train)[:,0]
y_test_pred = model.predict(X_test)[:,0]
y_test_pred[:5]
```

### Evalute performance

```python
from sklearn.metrics import r2_score, mean_squared_error

print(r2_score( y_train, y_train_pred ))
print(r2_score( y_test, y_test_pred ))

print(mean_squared_error( y_train, y_train_pred ))
print(mean_squared_error( y_test, y_test_pred ))
```
