```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from tqdm import tqdm

#from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

# Create data

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=1000,
                  centers=3,
                  cluster_std=1.3,
                  random_state=0)

X -= np.mean(X,axis=0)

%matplotlib inline

cmap = 'jet'
MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
scale = 3
X, y = make_blobs(n_samples=[300,200,100],
                  centers=[ [-scale,0], [scale,0], [0,scale] ],
                  cluster_std=0.4,
                  random_state=0)

X -= np.mean(X,axis=0)

MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=1, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
X, y = make_blobs(n_samples=500,
                  centers=3,
                  cluster_std=0.6,
                  random_state=1)

transformation = [[0.6, -0.6], [-0.4, 0.8]]
X = np.dot(X, transformation)

X -= np.mean(X,axis=0)

MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

```python
from sklearn.datasets import make_circles

X, y = make_circles(n_samples=500,
                    factor=0.5,
                    noise=0.05)

MAX = np.max(np.abs(X))*1.2
plt.scatter(X[:,0], X[:,1], c=y, s=5, cmap=cmap);
plt.xlim(-MAX,MAX)
plt.ylim(-MAX,MAX)
plt.gca().set_aspect('equal')
```

# Preprocess

### Oneihot encode

```python
from sklearn.preprocessing import OneHotEncoder

OHE = OneHotEncoder()
y = OHE.fit_transform(y.reshape(-1, 1)).toarray()
y
```

### Train/test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    stratify=y)
```

# Keras

### Create model

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()

#model.add(Dense(y.shape[1], input_dim=X.shape[1], activation='softmax'))

model.add(Dense(8, input_dim=X.shape[1], activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(y.shape[1], activation='softmax'))

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model.summary()
```

### Train model

```python
history = model.fit(X_train, y_train,
                    epochs=50,
                    batch_size=32,
                    validation_data=(X_test, y_test),
                    verbose=1)
```

```python
# Plot training and validation loss
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

# Plot training and validation accuracy
plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.ylim(0,1)
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```

### Print weights

```python
for i, layer in enumerate(model.layers):
    print(f"Weights of layer {i}:")
    print('Weights:')
    print(layer.get_weights()[0])
    print('Biasess:')
    print(layer.get_weights()[1])
```

### Predict model on the test set

```python
y_test_prob = model.predict(X_test)
y_test_prob[:5]
```

```python
y_test_pred = y_test_prob.argmax(axis=1)
y_test_pred[:5]
```

### Evalute performance

```python
from sklearn.metrics import  classification_report

print(classification_report(y_test.argmax(axis=1),
                            y_test_pred))
```

### Visualize decision boundaries

```python
nGrid = 200
a1 = np.linspace(-MAX,MAX,nGrid)
a2 = np.linspace(-MAX,MAX,nGrid)

A1, A2 = np.meshgrid(a1,a2)
A1 = A1.flatten()
A2 = A2.flatten()
A = np.vstack((A1,A2)).T

B = model.predict(A).argmax(axis=1)
B = B.reshape(nGrid,nGrid)
B = np.flipud(B)

extent = [-MAX,MAX,-MAX,MAX]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.gca()

ax.imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
ax.scatter(X[:,0], X[:,1], c=y.argmax(axis=1), s=20, cmap=cmap);
ax.set_aspect('equal')
```

# Visualize while training

```python
model2 = Sequential()

model2.add(Dense(8, input_dim=X.shape[1], activation='relu'))
model2.add(Dense(4, activation='relu'))
model2.add(Dense(y.shape[1], activation='softmax'))

model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# predict on grid
B = model2.predict(A).argmax(axis=1)
B = B.reshape(nGrid,nGrid)
B = np.flipud(B)

# plot
fig = plt.figure(figsize=(6,6))
ax = plt.gca()

ax.imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
ax.scatter(X[:,0], X[:,1], c=y.argmax(axis=1), s=20, cmap=cmap);
ax.set_title(f"epoch {0} (untrained)")
ax.set_aspect('equal')
plt.show()

for i in range(10):
    # train
    model2.fit(X_train, y_train,
               epochs=5,
               batch_size=32,
               validation_data=(X_test, y_test),
               verbose=0)
    
    # predict on grid
    B = model2.predict(A).argmax(axis=1)
    B = B.reshape(nGrid,nGrid)
    B = np.flipud(B)
    
    # plot
    fig = plt.figure(figsize=(6,6))
    ax = plt.gca()

    ax.imshow(B, interpolation='none', extent=extent,cmap=cmap, alpha=0.4)
    ax.scatter(X[:,0], X[:,1], c=y.argmax(axis=1), s=20, cmap=cmap);
    ax.set_title(f"epoch {(i+1)*5}")
    ax.set_aspect('equal')
    plt.show()
# end
```
