```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',100)
pd.set_option("display.precision", 4)

plt.style.use('dark_background')
```

# Entropy

```python
def entropy( P, b ):
    s = 0
    for p in P:
        if p > 0:
            s += -p*math.log(p,b)
        # end
    # end
    return s
# end
```

```python
p = np.zeros((100,2))
p[:,0] = np.linspace(0,1,100)
p[:,1] = 1 - p[:,0]

ents = [ entropy(p,2) for p in list(p) ]

plt.plot( p[:,0], ents )
```

```python
n = 3

a = 0.5
x = np.exp( -np.arange(n) / a )
x = x / x.sum()

print(entropy(x, 2))

sns.barplot(x=np.arange(n), y=x);
plt.ylim(0,1);
```

```python
x = np.random.normal(0.5,2.5,5000)
xlim = [-10,10]
bins = 100

counts, _ = np.histogram(x, bins=bins, range=xlim)
counts = counts / counts.sum()
print(entropy(counts,2))

sns.histplot(x,bins=bins,binrange=xlim);
#plt.ylim(0,1000);
```

# Cross entropy

```python
def cross_entropy( P, Q, b ):
    s = 0
    for i in range(len(P)):
        if Q[i] > 0:
            s += -P[i]*math.log(Q[i],b)
        # end
    # end
    return s
# end
```

```python
n = 3

a = 0.001
x = np.exp( -np.arange(n) / a )
x = x / x.sum()

b = 1
y = np.exp( -np.arange(n) / b )
y = y / y.sum()

print(cross_entropy(x, y, 2))

plt.plot(np.arange(n), x, 'b');
plt.plot(np.arange(n), y, 'r');
plt.ylim(0,1);
```
